```python
# beerId = list(beer_df.values.T.flatten())
# beerId
beerId = beer_df["ID"].tolist()
print(len(beerId))
beerId
```

```python
print(len(beerId))
base_url = f"https://sandbox-api.brewerydb.com/v2/beer/{beerId}/ingredients/?key={api_key}"
beer_data=[]
print("-----------------------------")
print("Beginning Data Retrieval     ")
print("-----------------------------")
# Create counters
record_count = 1


# Loop through all the beers in our list
for  beer in beerId:
        
   

    # Create endpoint URL with each beer
#     beer_url = url + "&q=" + beer
    
    # Log the url, record
    print("Processing Record %s  | %s" % (record_count, beer))

    # Add 1 to the record count
    record_count += 1

    # Run an API request for each of the beers
    try:
        beer_fact = requests.get(base_url).json()
        beer_data.append(beer_fact['data'])
        
#         Skip if the record of the beer is incomplete
    except:
        print("Beer not found. Skipping...")
        pass
        
              
# Indicate that Data Loading is complete 

print("----------------------------")
print("Data Retrieval Complete     ")
print("----------------------------")
```
